## 1 Modeling Volatility and VaR

### 1.1 Historic VaR

#### (a) Plot the historic-based VaR $\tilde{r}^{VaR, 0.05}_{t,historic}$ over time

```python
Historic_VaR = df['Excess SPY Returns'].shift(1).expanding(min_periods = 60).quantile(0.05)
Historic_VaR.plot()
plt.title('Historic VaR')
plt.ylabel('VaR')
plt.xlabel('Date')
plt.show()
```

#### (b) Calculate the frequency of periods in which $\tilde{r} < \tilde{r}^{VaR, 0.05}_{t,historic}$

```python
print(
    f"The frequency = {(df['Excess SPY Returns'] < Historic_VaR).sum()/len(Historic_VaR.dropna())*100:.2f}%")
```

#### (c) What drawbacks do you see in this historic VaR?

- Historic VaR puts equal weights on ALL past data from 1930s, while more recent data should be more relevant to the markets than data from longer period. Most of the past data may not accurately capture recent trends or structural changes in the market, thus being less relevant.
- Even if the historic VaR is improved to only look back on more recent data, it not may be able to find sufficient data points to make statistical sense.

1. Backward Looking: Looks at only the historical data while estimating the losses in the future. Availability of sufficient and relevant data becomes a challenge. 


2. Ghosting Effect: Equal Weights on past observations that may never be relevant again. For example, low returns during 2008 (which may not be relevant anymore) are given the same weight as the most recent observations


3. Slow to React: Less Weight on most recent observations. A recent significant loss still does not impact historical VaR because it has the same weight as the other observations

### 1.2 Volatility

#### (a) For these two methods, plot $\sigma_t$.

```python
vol = pd.DataFrame()
vol = np.sqrt((df['Excess SPY Returns'] **
              2).shift(1).expanding(60).mean().to_frame('Expanding'))
vol['Rolling'] = np.sqrt(
    (df['Excess SPY Returns']**2).shift(1).rolling(60).mean())
# vol['vol_expanding'] = df['Excess SPY Returns'].shift(1).expanding(60).std()
# vol['vol_rolling'] = df['Excess SPY Returns'].shift(1).rolling(60).std()
vol.plot()
plt.title('Estimated Volatility')
plt.xlabel('Date')
plt.ylabel('Vol')
plt.show()
```

#### (b) For each method, calculate the 5th percentile, 1-month-ahead VaR using the formula
$$\tilde{r}^{VaR, 0.05}_{t,vol} = −1.65 \sigma_t$$

```python
VaR = (vol * -1.65)
VaR.rename(columns={'vol_expanding': 'Expanding',
           'vol_rolling': 'Rolling'}, inplace=True)
VaR
```

#### (c) Calculate the frequency of periods in which $\tilde{r} < \tilde{r}^{VaR, 0.05}_{t,vol}$

```python
print(
    f"For expanding VaR, frequency = {(df['Excess SPY Returns'] < VaR['Expanding']).sum()/len(VaR.dropna())*100:.2f}%")
print(
    f"For rolling VaR, frequency = {(df['Excess SPY Returns'] < VaR['Rolling']).sum()/len(VaR.dropna())*100:.2f}%")
```

#### (d) Compare the historic VaR and two vol-based VaR

```python
VaR['Historic'] = Historic_VaR
VaR.plot()
plt.title('Comparing VaR Estimates')
plt.xlabel('Date')
plt.ylabel('VaR')
plt.show()
```

 - The rolling vol-based VaR is much more volatile than the other two, while expanding vol-based VaR and Historic VaR have very similar trends.
 - From the frequency comparison, we could see that the rolling vol-based VaR has frequency 4.95% closest to 5%, indicating that this VaR estimate has the best performance in sample among three.哪个离5%最近最好，因为是5%quantile VaR并且rolling count most recent observations --> more accurate

## 3 Estimate Underperformance

### 3.1 Summary Statistics

#### (a) Report of statistics

```python
stats_table = pd.DataFrame(
    index=pd.MultiIndex.from_tuples([('levels', 'SPX'), ('levels', 'RiskFree'), (
        'levels', 'Excess SPX'), ('logs', 'SPX'), ('logs', 'RiskFree'), ('logs', 'Excess SPX')]),
    columns=pd.MultiIndex.from_tuples([('1965-1999', 'mean'), ('1965-1999', 'vol'), ('2000-2023', 'mean'), ('2000-2023', 'vol'), ('1926-2023', 'mean'), ('1926-2023', 'vol')]))


stats_table[('1965-1999', 'mean')] = ((df.loc['1965':'1999']).mean()* 12).tolist() + ((np.log(1+df.loc['1965':'1999'])).mean()*12).tolist()
stats_table[('1965-1999', 'vol')] = ((df.loc['1965':'1999']).std() * np.sqrt(12)).tolist() + ((np.log(1+df.loc['1965':'1999'])).std() * np.sqrt(12)).tolist()
stats_table[('2000-2023', 'mean')] = ((df.loc['2000':'2023']).mean()* 12).tolist() + ((np.log(1+df.loc['2000':'2023'])).mean()*12).tolist()
stats_table[('2000-2023', 'vol')] = ((df.loc['2000':'2023']).std() * np.sqrt(12)).tolist() + ((np.log(1+df.loc['2000':'2023'])).std() * np.sqrt(12)).tolist()
stats_table[('1926-2023', 'mean')] = (np.mean(df) * 12).tolist() + (np.mean(np.log(1+df))*12).tolist()
stats_table[('1926-2023', 'vol')] = (df.std()* np.sqrt(12)).tolist() + ((np.log(1+df)).std() * np.sqrt(12)).tolist()

display(stats_table.style.format('{:,.2%}'))
```

#### (b) Comment on how the full-sample return stats compare to the sub-sample stats, and how the level stats compare to the log stats.

- Market return and risk-free return both have lower mean and higher vol in full-sample than in sub-sample, market excess return have higher mean and higher vol in full-sample.
- Market level returns are significantly higher than market log returns in full-sample and sub-sample, while market vol, risk-free mean and vol are very close in level stats and log stats.

<font color = "red">
    Full sample estimates are significantly different for the excess returns, this is because of the low risk-free rate during that period. The log returns show lower mean returns as expected.

### 3.2 $$Pr[R^M_{t,t+h} < R^f_{t,t+h}]$$ <div style="text-align: right">

```python
def prob(mu, sigma, h):
    return norm.cdf(-np.sqrt(h)*mu/sigma)
#market return < risk free  -->  excess return < 0  --> 0-mu * sqrt(time)/sigma

mu = stats_table[('1965-1999', 'mean')][('logs', 'Excess SPX')]
sigma = stats_table[('1965-1999', 'vol')][('logs', 'Excess SPX')]

print('Probability of SPX underperforming RF in the next 15 years after 1999: {:,.2%}'.
      format(prob(mu=mu, sigma=sigma, h=15)))

print('Probability of SPX underperforming RF in the next 30 years after 1999: {:,.2%}'.
      format(prob(mu=mu, sigma=sigma, h=30)))
```

```python
h = list(range(1, 31))
probs = pd.DataFrame([prob(mu, sigma, x) for x in h], columns=[
                     'Probability of Underperformance'], index=h)
probs.plot()
plt.show()
```
