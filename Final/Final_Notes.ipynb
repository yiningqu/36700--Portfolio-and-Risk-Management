### Time-series test vs. Cross-sectional test
- time series test: fits the asset, insist the factor premia for each factor are sample averages, only requires Alpha, R^2 not important 
- cross sectional test: fits the asset looser, let factor premia (lambda) be whatever value to fit the model better, cares about R^2

# Autoregressive model (AR)
serial correlation in equity is small, in risk free rate is very high

$$cov[r_t,r_{t+i}] = ρ^{i}σ^{2}$$

- ρ = 1, stdarnd dev is linear in cumulation: $std[r_{t,t+h}] = hσ$
- ρ = 0, variance is linear in cumulation: $var[r_{t,t+h}] = hσ^{2}$, $std[r_{t,t+h}] = \sqrt{h}σ$
    - 因为each period is independent with std dev σ, together is hσ
- ρ = -1, stdarnd dev is linear in cumulation: $var[r_{t,t+h}] = 0$

### riskless bond: 只有起始和结束点的价格是fixed，中间的price is uncertain
- if $r_{0,t}$ is high, then $r_{t,10}$ must be relatively low
- 所以有negative autocorrelated return: 
    - $cov[r_t,r_{t+1}] < 0$
    - $cov[r_{t,t+h},r_{t+h,T}] = -1 $
    - $var[r_{t,T}] = 0$ in long run


### Sharpe Ratio
- ρ = 1, $SR(r_{t,t+h}) = SR(r_t)$
- |ρ| < 1, $SR(r_{t,t+h}) > SR(r_t)$
    - 时间越长，SR越大
    - 可以理解为和different asset一样，因为period之间有low autocorrelation，所以时间越长，risk reuduce越多
- ρ = 0, $SR(r_{t,t+h}) = \sqrt{h}SR(r_t)$

### CIP (Covered Interest Parity):
$$f^s_t - s_t = r^{f,$}_{t,t+1} - r^{f,€}_{t,t+1} $$

$$E[\frac{F^s_t}{S_t}] R^{f,€}_{t,t+1} = R^{f,$}_{t,t+1}$$
- follow Law of One Price
- 左边式子=log forward premium 
- 右边式子=interest rate differential
    - invest EU risk free rate at time t+1 + 
    - buy forward contract at time t to transfer back to USD at time t+1 
    - = invest USD risk free rate at time t+1 $$


### UIP (Uncovered interest parity):
$$E[\frac{S_{t+1}}{S_t}]  = \frac{R^{f,$}_{t,t+1}}{R^{f,€}_{t,t+1}}$$

- 左边式子=exchange growth rate
    - invest in USD risk free rate at time t=
    - invest in EURO risk free rate at time t +
    - convert payoff back to USD at time t+1
#### depreciation of USD (S增加） offset IR premium over EURO （risk free rate增加）

$$lnE[S_{t+1}] - s_t  = r^{f,$}_{t,t+1} - r^{f,€}_{t,t+1} = f^s_t - s_t$$

#### forward rate is best predictor of future spot rate (exchange rate)
### CIP & UIP 区别：
    - CIP holds when no arbitrage 有condition
    - UIP is a theory 无condition
    - UIP is risk neutral to FX rate （assume FX risk is not priced)
    - neither implies the other. CIP is a no-arbitrage relationship of forward rates vs spot rates. UIP is a hypothesis that FX risk does not earn a premium; in other words, expected excess returns of FX are zero. The theory does not imply the no-arbitrage relationship, nor vice-versa. It is true that sometimes we use the CIP relationship to insert into UIP to write it in a different form, but this is not enough to prove UIP; it is simply restating it in terms of forward rates.
